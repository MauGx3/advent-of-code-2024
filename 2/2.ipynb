### Part One

Fortunately, the first location The Historians want to search isn't a long walk from the Chief Historian's office.

While the [Red-Nosed Reindeer nuclear fusion/fission plant](https://adventofcode.com/2015/day/19) appears to contain no sign of the Chief Historian, the engineers there run up to you as soon as they see you. Apparently, they _still_ talk about the time Rudolph was saved through molecular synthesis from a single electron.

They're quick to add that - since you're already here - they'd really appreciate your help analyzing some unusual data from the Red-Nosed reactor. You turn to check if The Historians are waiting for you, but they seem to have already divided into groups that are currently searching every corner of the facility. You offer to help with the unusual data.

The unusual data (your puzzle input) consists of many _reports_, one report per line. Each report is a list of numbers called _levels_ that are separated by spaces. For example:

    7 6 4 2 1
    1 2 7 8 9
    9 7 6 2 1
    1 3 2 4 5
    8 6 4 4 1
    1 3 6 7 9

This example data contains six reports each containing five levels.

The engineers are trying to figure out which reports are _safe_. The Red-Nosed reactor safety systems can only tolerate levels that are either gradually increasing or gradually decreasing. So, a report only counts as safe if both of the following are true:

* The levels are either _all increasing_ or _all decreasing_.
* Any two adjacent levels differ by _at least one_ and _at most three_.

In the example above, the reports can be found safe or unsafe by checking those rules:

* `7 6 4 2 1`: _Safe_ because the levels are all decreasing by 1 or 2.
* `1 2 7 8 9`: _Unsafe_ because `2 7` is an increase of 5.
* `9 7 6 2 1`: _Unsafe_ because `6 2` is a decrease of 4.
* `1 3 2 4 5`: _Unsafe_ because `1 3` is increasing but `3 2` is decreasing.
* `8 6 4 4 1`: _Unsafe_ because `4 4` is neither an increase or a decrease.
* `1 3 6 7 9`: _Safe_ because the levels are all increasing by 1, 2, or 3.

So, in this example, _`2`_ reports are _safe_.

Analyze the unusual data from the engineers. _How many reports are safe?_

```python
with open('input.txt', 'r') as f:
    data = f.readlines()
```

```python
lines = [line.split() for line in data]
lines
```

```python
converted_lines = [[int(number) for number in line] for line in lines]
converted_lines
```

```python
valid_lines = []
invalid_lines = []

for line in converted_lines:
    line_distances = []

    for i in range(len(line) - 1):
        distance = line[i] - line[i + 1]
        line_distances.append(distance)

    if line_distances:
        valid_distances = all(abs(n) >= 1 and abs(n) <= 3 for n in line_distances)
        all_positive = all(n > 0 for n in line_distances)
        all_negative = all(n < 0 for n in line_distances)
        result = (all_positive or all_negative) and valid_distances

        if result is True:
            valid_lines.append(line)
        else:
            invalid_lines.append(line)
```

```python
def rule_check(lines):
    line_check = {
        "True": {"index": [], "count": 0},
        "False": {"index": [], "count": 0},
        "Positive": {"index": [], "count": 0},
        "Negative": {"index": [], "count": 0},
        "Zero": {"index": [], "count": 0},
    }

    for index, line in enumerate(lines):
        valid_distance = str(line[0])
        posneg = line[1]
        if valid_distance in line_check:
            line_check[valid_distance]["index"].append(index)
            line_check[valid_distance]["count"] += 1
        if posneg in line_check:
            line_check[posneg]["index"].append(index)
            line_check[posneg]["count"] += 1

    return line_check
```

```python
dampened_lines = []
for index, line in enumerate(lines_with_rules):
    dampener = []
    result = rule_check(line)
    print(f"Line {index+1}: {line}, \n\n Result: {result}")
    # print('\n')

    # needs dampener
    if result["True"]["count"] > 0 and result["False"]["count"] > 0:
        print(result["True"]["index"])
        print(result["False"]["index"])
    # doesn't need dampener
    else:
        print("Safe")
        dampener.append(True)
    # dampener won't work
    if len(result["True"]["index"]) > 1 and len(result["False"]["index"]) > 1:
        print("Unsafe")
        dampener.append(False)
    # dampener works
    else:
        print("Safe")
        dampener.append(True)

    if result["Positive"]["count"] > 0 and result["Negative"]["count"] > 0:
        print(result["Positive"]["index"])
        print(result["Negative"]["index"])
    else:
        print("Safe")
        dampener.append(True)
    if len(result["Positive"]["index"]) > 1 and len(result["Negative"]["index"]) > 1:
        print("Unsafe")
        dampener.append(False)
    else:
        print("Safe")
        dampener.append(True)

    if (
        result["False"]["count"] > 0
        and result["Positive"]["count"] > 1
        and result["Negative"]["count"] > 1
    ):
        print("Unsafe")
        dampener.append(False)

    print(dampener)
    dampened_lines.append(dampener)
    print("\n")
```

```python
def is_valid_sequence(line):
    """
    Check if a sequence is valid according to ALL puzzle rules:
    1. Differences must be consistently positive or negative
    2. Absolute differences must be between 1 and 3

    Args:
        line (list): List of numbers to check

    Returns:
        bool: True if sequence follows all rules
    """
    # Calculate all differences between adjacent numbers
    differences = []
    for i in range(len(line) - 1):
        difference = line[i] - line[i + 1]
        differences.append(difference)

    if not differences:  # Handle empty sequences
        return True

    # Check three conditions:
    # 1. All differences must have absolute value between 1 and 3
    valid_distances = all(abs(n) >= 1 and abs(n) <= 3 for n in differences)

    # 2. All differences must be positive
    all_positive = all(n > 0 for n in differences)

    # 3. All differences must be negative
    all_negative = all(n < 0 for n in differences)

    # Final result: must satisfy the distance constraint AND be either all positive or all negative
    return (all_positive or all_negative) and valid_distances
```

```python
def check_safety(line):
    """
    Check if a sequence is safe according to the puzzle rules.
    A sequence is safe if it's either:
    1. Already consistently increasing or decreasing
    2. Can be made consistently increasing or decreasing by removing one number

    Args:
        line (list): List of numbers to check

    Returns:
        bool: True if the sequence is safe, False otherwise
    """
    # First check if sequence is already valid
    if is_valid_sequence(line):
        return True

    # If not valid, try removing each number one at a time
    for i in range(len(line)):
        # Create a new sequence without the current number
        test_sequence = line[:i] + line[i + 1 :]

        # If this new sequence is valid, we've found a solution
        if is_valid_sequence(test_sequence):
            return True

    # If we get here, no valid solution was found
    return False
```
